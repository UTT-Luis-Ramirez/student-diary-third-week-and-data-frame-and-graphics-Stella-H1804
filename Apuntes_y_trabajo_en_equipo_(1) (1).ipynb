```python
population_size = len(data)
print("El tamaño de la población es: ", population_size)
```

###MUESTREO
Creación de una muestra aleatoria de 50 registros

```python
sample_size = 50
sample = data.sample(n = sample_size, random_state=7)
```

###REVISIÓN DE LA MUESTRA

```python
display(sample)
```

```python
sample_size = len(sample)
print(sample_size)
```

###MUESTREO ALEATORIO SISTEMATICO
Definición de intervalo k (tamaño de la población= entre el tamaño de la muestra deseada)

```python
sample_size = 100
interval_k = population_size // sample_size
```

###SELECCIÓN DE MUESTRA SISTEMATICA

```python
systematic_sample = data.iloc[::interval_k]
```

```python
print("Muestra obtenida por muestreo sistematico")
display(systematic_sample)
```

###DIVIDIR POBLACIÓN POR UN FACTOR
